```python
def identify_spikes(arr):
    equality = np.array(arr) == np.array([0 for i in range(len(arr))])
    indices = [i for i in range(len(equality)) if not equality[i]]
    return indices
```

```python
def has_repeat(seq, diff_profile, repeat, smin=3):
    is_repeat = []
    spikes = identify_spikes(diff_profile)
    print("spikes: " + str(spikes))
    for spike in spikes:
        for i in range(spike):
            overlap_len = spike - i
            if seq[i:spike] == repeat[:overlap_len] or seq[i:spike] == repeat[len(repeat)-overlap_len:]:
                is_repeat.append(True)
                break
    return is_repeat
```
